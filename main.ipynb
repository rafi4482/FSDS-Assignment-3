1. Inventory Price Calculator

```python
def calculate_inventory_summary(prices):
    return sum(prices),sum(prices)/len(prices)

print(calculate_inventory_summary([120, 250, 399, 150]))
```

2. Unique Visitor Tracker

```python
def count_unique_visitors(visitor_list):
    unique_count = 0
    
    for i in range(len(visitor_list)):
        seen_before = False
        
        for j in range(i):
            if visitor_list[i] == visitor_list[j]:
                seen_before = True
                break
        
        if not seen_before:
            unique_count += 1

    return unique_count

print(count_unique_visitors([101, 205, 101, 310, 205, 550]))
```

3. Product Category Counter

```python
def count_categories(products):
    products_dict = {}
    for product in products:
        products_dict[product] = products.count(product)

    return products_dict


print(count_categories(["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]))
```

4. Temperature Analytics

```python
def analyze_temperatures(temp_list):
    high_temp = []
    normal_temp = []

    for temp in temp_list:
        if temp > 38:
            high_temp.append(temp)
        else:
            normal_temp.append(temp)

    return high_temp,normal_temp

print(analyze_temperatures([36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]))

```

5. Menu Price Lookup

```python
def get_price(menu, item):
    
    for key,value in menu.items():
        if key == item:
            return value
    return "Item not found"

print(get_price({"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50},"Muffin"))
```

6. Student Grading System

```python
def calculate_grade(score):
    if score >= 85:
        return "A+"
    elif 70 <= score <= 84:
        return "A"
    elif 33 <= score <= 69:
        return "A-"
    else:
        return "F"

def grade_students(student_dict):
    grades = {}
    for name, score in student_dict.items():
        grades[name] = calculate_grade(score)
    return grades

result = grade_students({"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69})
print(result)
```
